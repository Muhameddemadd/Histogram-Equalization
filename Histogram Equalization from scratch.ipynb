```python
def Histogram_equalization(img_path):
    img = cv2.imread(img_path)
    img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    flat = img.flatten()
    histogram = np.zeros(256)
    Csum=[]
    x=0

    for pixel in img:
        histogram[pixel] += 1

    for i in range(0,len(histogram)):
        x+=histogram[i]
        Csum.append(x)

    minimum = min(Csum)
    maximum = max(Csum)

    nj = (Csum - minimum) * 255
    N = maximum - minimum
    Csum = nj / N
    Csum = Csum.astype('uint8')
    output = Csum[flat]
    img_new = np.reshape(output, img.shape)

    return plt.imshow(img_new)
```

```python
Histogram_equalization("image-2.png")
```
